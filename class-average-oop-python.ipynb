```python
# ============================================
# Class Average Program - OOP in Python
# CIT 2228 - Computer Programming II
# ============================================

class Student:
    """Represents a single student with a name and list of grades."""

    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # list of numeric grades

    def get_average(self):
        """Calculate and return this student's average grade."""
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

    def __str__(self):
        avg = self.get_average()
        return f"{self.name}: Grades = {self.grades}, Average = {avg:.2f}"


class ClassAverage:
    """Manages a collection of students and computes the class average."""

    def __init__(self):
        self.students = []  # empty list to hold Student objects

    def add_student(self, student):
        """Add a Student object to the class."""
        self.students.append(student)

    def calculate_class_average(self):
        """Calculate the average of all students' averages."""
        if len(self.students) == 0:
            return 0
        total = sum(student.get_average() for student in self.students)
        return total / len(self.students)

    def display_results(self):
        """Display each student's info and the overall class average."""
        print("\n========== CLASS RESULTS ==========")
        for student in self.students:
            print(student)
        print("------------------------------------")
        print(f"Class Average: {self.calculate_class_average():.2f}")
        print("====================================")


# ============================================
# MAIN PROGRAM
# ============================================

def main():
    classroom = ClassAverage()

    n = int(input("Enter the number of students: "))

    for i in range(n):
        print(f"\n--- Student {i + 1} ---")
        name = input("Enter student name: ")

        grades_input = input("Enter grades separated by spaces (e.g. 80 75 90): ")
        grades = [float(g) for g in grades_input.split()]

        student = Student(name, grades)
        classroom.add_student(student)

    classroom.display_results()


main()
```
